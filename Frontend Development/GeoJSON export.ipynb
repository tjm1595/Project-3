```python
import sqlite3
import json
```

```python
connection = sqlite3.connect("C:/Users/redye\Downloads/FPA_FOD_20170508.sqlite")
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("SELECT FIRE_YEAR, DISCOVERY_DOY, DISCOVERY_TIME, CONT_DATE, CONT_DOY, CONT_TIME, FIRE_SIZE, LATITUDE, LONGITUDE, STATE, STAT_CAUSE_CODE, STAT_CAUSE_DESCR FROM Fires WHERE FIRE_SIZE > 1000 AND (STATE = 'CO' OR STATE = 'MT' OR STATE = 'WY')")
```

```python
rows = cursor.fetchall()
```

```python
rows
```

```python
features = []
```

```python
for row in rows:
    # Define the properties for the feature
    properties = {
        'FIRE_YEAR': row[0],
        'DISCOVERY_DOY': row[1],
        'DISCOVERY_TIME': row[2],
        'CONT_DATE': row[3],
        'CONT_DOY': row[4],
        'CONT_TIME': row[5],
        'FIRE_SIZE': row[6],
        'STATE': row[9],
        'STAT_CAUSE_CODE': row[10],
        'STAT_CAUSE_DESCR': row[11]
    }

    # Define the geometry for the feature
    geometry = {
        'type': 'Point',
        'coordinates': [row[8], row[7]]  # Longitude, Latitude
    }

    # Create the GeoJSON feature
    feature = {
        'type': 'Feature',
        'properties': properties,
        'geometry': geometry
    }

    # Add the feature to the features list
    features.append(feature)
```

```python
geojson = {
    'type': 'FeatureCollection',
    'features': features
}
```

```python
with open('output3.geojson', 'w') as file:
    json.dump(geojson, file, indent=2)
```
